```python


arr1 = np.array([1, 2, 3, 4])

arr2 = np.array([5, 6, 7, 8])
arr = np.concatenate((arr1,arr2),axis=0)
print(arr)
```

```python


arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
arr3 = np.array([[9, 10], [11, 12]])

# Concatenate along axis 0 (rows)
result_axis_0 = np.concatenate((arr1, arr2,arr3), axis=0)

# Concatenate along axis 1 (columns)
result_axis_1 = np.concatenate((arr1, arr2,arr3), axis=1)
print(result_axis_0)
print(result_axis_1)
```

```python


# Two 3D arrays
arr1 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
arr2 = np.array([[[9, 10], [11, 12]], [[13, 14], [15, 16]]])

# Concatenate along axis=2 (columns within each slice)
result_axis_2 = np.concatenate((arr1, arr2), axis=2)
print(result_axis_2)
```
